1. Make sure your data is BIDS compliant: https://bids.neuroimaging.io/
2. run fMRIprep https://fmriprep.org/en/stable/# to get GM and WM probability maps
3. This script extracts brains

```python
import ants
from matplotlib import pyplot as plt
import os
```

```python
# anat_fn = '~/data/ABIDE/Derivatives/fmriprep3/sub-0050002/anat/sub-0050002_desc-preproc_T1w.nii.gz'
# gm_mask = '~/data/ABIDE/Derivatives/fmriprep3/sub-0050002/anat/sub-0050002_label-GM_probseg.nii.gz'
# wm_mask = '~/data/ABIDE/Derivatives/fmriprep3/sub-0050002/anat/sub-0050002_label-WM_probseg.nii.gz'
anat_fn = 'Data/output/ds001796/sub-201/anat/sub-201_desc-preproc_T1w.nii.gz'
gm_mask = 'Data/output/ds001796/sub-201/anat/sub-201_label-GM_probseg.nii.gz'
wm_mask = 'Data/output/ds001796/sub-201/anat/sub-201_label-WM_probseg.nii.gz'
```

```python
t1 = ants.image_read(anat_fn) # filename for raw anatomical 
gm = ants.image_read(gm_mask) # gray matter probability map
wm = ants.image_read(wm_mask) # white matter probability map
brain_mask = gm+wm # combine GM and WM 

t1.plot_ortho(brain_mask,flat=True,xyz_lines=False,orient_labels=False,figsize=5.0,overlay_alpha=.5,overlay_cmap='hot')
```

```python
brain  = t1.copy() # Make a copy to not overwrite the original
brain[brain_mask<.9] = 0 # Zero out the non-brain stuff
brain.plot_ortho(flat=True,xyz_lines=False,orient_labels=False,figsize=5.0) # Check if it worked
```

```python
plt.hist(brain.numpy().flatten()); # check values 
```

```python
brain_tiny = brain.resample_image(resample_params=(64,64,64), use_voxels=True, interp_type=4) # Make sure all brains are same size
brain_tiny.plot_ortho(flat=True,xyz_lines=False,orient_labels=False,figsize=5.0) 
```

```python
brain.to_filename('brain.nii') # Save extracted brain
```
