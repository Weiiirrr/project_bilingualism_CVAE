```python
data = data[data['lan_know'].notna()]
data['lan_know'].value_counts(dropna=False)
```

```python
#ds, participant_id, bilingualism(0-no, 1-yes)
#age, sex(F,M), scanner
#aoa/aoa_group(EB, LB), num_lan, sec_lan_proficiency(0-1), sec_lan_exposure(0-1)
```

```python
data = data.copy()
data.loc[:,'participant_id'] = data.loc[:,('src_subject_id')]
```

```python
data.loc[data['lan_know']=='1','bilingualism'] = 0
data.loc[data['lan_know']!='1','bilingualism'] = 1
```

```python
data['interview_age'] = pd.to_numeric(data['interview_age'])
print(data['interview_age'].max())
print(data['interview_age'].min())
data['interview_age'].hist(bins=5).set_xticks([0, 200, 400, 600, 800, 1000, 1200, 1400])
plt.show()
```

```python
data.loc[:, 'age'] = data.loc[:,'interview_age'].astype(float)/12
# data['age'].mean() #mean of age is 60.36
# data['sex'] = data['sex']
```

```python
data.age.describe()
```

```python
data.lan521 = pd.to_numeric(data.lan521)
print(data['lan521'].describe())
print(data['lan521'].max())
print(data['lan521'].min())
data['lan521'].hist()
plt.show()

data['lan521'].hist(range=[0,80])
plt.show()
```

```python
data.loc[:, 'aoa'] = data.loc[:, 'lan521'] #'lan521': Age when you first: Language 2: Began acquiring Language 2:
data.aoa = data.aoa.replace(999, np.nan)
print(data['aoa'].describe())
data['aoa'].hist()
```

```python
data.loc[data['aoa'].astype(float)>12, 'aoa_group'] = 'LB'
data.loc[data['aoa'].astype(float)<=12, 'aoa_group'] = 'EB'
```

```python
data['aoa_group'].value_counts(dropna=False)
```

```python
data['num_lan'] = data['lan_know']
data['num_lan'] = data['num_lan'].astype(int)
```

```python
#average all proficiency measures (speak, read, listen)
# legend['lan1_pro'] = (df['lan611'].astype(float)+df['lan612'].astype(float)+df['lan613'].astype(float))/3
# legend['lan3_pro'] = (df['lan631'].astype(float)+df['lan632'].astype(float)+df['lan633'].astype(float))/3
# legend['lan4_pro'] = (df['lan641'].astype(float)+df['lan642'].astype(float)+df['lan643'].astype(float))/3
data.loc[:,'sec_lan_proficiency'] = (data['lan621'].astype(float)+data['lan622'].astype(float)+data['lan623'].astype(float))/30
```

```python
data.loc[:, 'sec_lan_exposure'] = data['lan32'].astype(float)
```
