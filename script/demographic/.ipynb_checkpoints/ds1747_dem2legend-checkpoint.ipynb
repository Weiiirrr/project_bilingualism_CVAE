```python
import pandas as pd
import os
```

```python
data = pd.read_csv(os.path.expanduser('~/Data/Bilingualism_CVAE/raw/ds1747/participants.tsv'), sep="\t")
data
```

```python
#ds, participant_id, bilingualism(0-no, 1-yes)
#age, sex(F,M), scanner
#aoa/aoa_group, num_lan, sec_lan_proficiency(0-1), sec_lan_exposure(0-1)
```

```python
data = data.copy()
data.loc[:,'num_lan'] = data.loc[:,'num_lang']
data.loc[data.group=='MC','bilingualism']=0
data.loc[data.group=='EB','bilingualism']=1
data.loc[data.group=='LB','bilingualism']=1
```

```python
#use the less proficient language as the second language
data.loc[:, 'sec_lan_proficiency'] = data.loc[(data.group == 'LB')|(data.group == 'EB'), ['eng_proficiency_score', 'sp_proficiency_score']].min(axis=1)/10
data.loc[data.group=='EB', 'aoa_group'] = 'EB'
data.loc[data.group=='LB', 'aoa_group'] = 'LB'
data.loc[:, 'ds'] = 'ds1747'
data.loc[:, 'scanner'] = 'Erlangen'
```

```python
data['group'].value_counts()
```

```python
legend = data.loc[:, ['participant_id', 'bilingualism', 'age', 'sex', 
              'aoa_group', 'num_lan', 'sec_lan_proficiency',
              'ds', 'scanner']]
```
